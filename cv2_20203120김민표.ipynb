```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
```

```python
img_path='/c.jpg'
```

```python
img=cv2.imread(img_path)
cv2_imshow(img)
```

```python
img1=cv2.resize(img,dsize=(0,0),fx=2,fy=2)
cv2_imshow(img1)
```

```python
img2=cv2.resize(img,dsize=(0,0),fx=0.5,fy=0.5)
cv2_imshow(img2)
```

```python
px=img[100,100] #B,G,R
print(px)
```

```python
crop_img = img[100:200,20:300]
cv2_imshow(crop_img)
```

```python
cv2.imwrite('/content/drive/MyDrive/c.jpg',img)
cv2.imwrite('/content/drive/MyDrive/c1.jpg',img1)
cv2.imwrite('/content/drive/MyDrive/c2.jpg',img2)
cv2.imwrite('/content/drive/MyDrive/c3.jpg',crop_img)
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_8U, ksize=3)
cv2_imshow(laplacian)
```

```python
sobel = cv2.Sobel(img, cv2.CV_8U, 1, 0, 3)
cv2_imshow(sobel)
```

```python
canny = cv2.Canny(img, 100, 255)
cv2_imshow(canny)
```
